```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
```

### Load Data

```python
fert = pd.read_csv(".\Data\gapminder_total_fertility.csv", index_col=0)
```

```python
continents = pd.read_csv(".\Data\continents.csv", sep=';',index_col=1)
continents
```

```python
life = pd.read_excel(".\Data\gapminder_lifeexpectancy.xlsx", index_col=0)
life.head()
```

```python
population = pd.read_excel(".\Data\gapminder_population.xlsx", index_col=0)
population.head()
```

### Convert Datatyps

```python
fert.columns = fert.columns.astype(int)
life.columns = life.columns.astype(int)
population.columns = population.columns.astype(int)
```

```python
type(fert.columns),type(life.columns),type(population.columns)
```

### Reindex Dataframes

```python
fert.index.name = 'country'
life.index.name = 'country'
population.index.name = 'country'
#continents.index.name ='country'
```

```python
type(fert.index[1]),type(life.index[1]),type(population.index[1]), type(continents.index[1])
```

```python
fert = fert.merge(continents, left_index=True, right_index=True)
```

```python
fert
```

```python
population=population.sort_values(by=population.columns[-1], ascending=False)
population.iloc[:,-1].head(20).plot.bar()
plt.title('Population in 2015_Top20')
```

```python
fert = fert.reset_index()
life = life.reset_index()
population = population.reset_index()
```

```python
type(fert.index[0]), type(life.index[0]), type(population.index[0])
```

```python
fert.head()
```

```python
fert = fert.melt(id_vars=['country', 'continent'], var_name='year', value_name='fertility_rate')
life = life.melt(id_vars='country', var_name='year', value_name='life_expectancy')
population = population.melt(id_vars='country', var_name='year', value_name='total_population')
```

```python
df=fert.merge(population)
df= df.merge(life)
```

```python
def normalize(x):
    return (x-x.min()) / (x.max()-x.min()) 
```

```python
df['total_population_z'] = normalize(df['total_population']).round(4)
```

```python
for yr in df_subset['year'].unique():
    df_year= df_subset.loc[df['year'] == yr]
    sp = sns.scatterplot(x='life_expectancy', y='fertility_rate',hue="continent", size='total_population',data=df_year, alpha=0.6)
    sns.move_legend(sp, "upper left", bbox_to_anchor=(1, 1))
    sp.set(ylim=(0,9))
    sp.set(xlim=(0,90))
    sp.set_title('{}'.format(yr))
    plt.savefig('image_png/{}.png'.format(yr))
    plt.clf()
```

```python
import imageio

images = []

for yr in df_subset['year'].unique():
    filename = 'image_png/{}.png'.format(yr)
    images.append(imageio.imread(filename))

imageio.mimsave('output1.gif', images, fps=20)
```
